```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['input_1'] = [0,0,1,1]
or_data['input_2'] = [0,1,0,1]
or_data['output'] = [0,1,1,1]
```

```python
or_data
```

```python
sns.scatterplot(x=or_data['input_1'],y=or_data['input_2'],hue=or_data['output'],s=200)
```

```python
and_data['input_1'] = [0,0,1,1]
and_data['input_2'] = [0,1,0,1]
and_data['output'] = [0,0,0,1]
```

```python
and_data
```

```python
sns.scatterplot(x=and_data['input_1'],y=and_data['input_2'],hue=and_data['output'],s=200)
```

```python
xor_data['input_1'] = [0,0,1,1]
xor_data['input_2'] = [0,1,0,1]
xor_data['output'] = [0,1,1,0]
```

```python
xor_data
```

```python
sns.scatterplot(x=xor_data['input_1'],y=xor_data['input_2'],hue=xor_data['output'],s=200)
```

```python
from sklearn.linear_model import Perceptron

clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(and_data.iloc[:,0:2].values,and_data.iloc[:,-1].values)
clf2.fit(or_data.iloc[:,0:2].values,or_data.iloc[:,-1].values)
clf3.fit(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values)
```

```python
clf1.coef_
```

```python
clf1.intercept_
```

```python
x = np.linspace(-1,1,5)
y = -x + 1
```

```python
plt.plot(x,y)
sns.scatterplot(x=and_data['input_1'],y=and_data['input_2'],hue=and_data['output'],s=200)
```

```python
clf2.coef_
```

```python
clf2.intercept_
```

```python
x1=np.linspace(-1,1,5)
y1=-x+0.5
```

```python
plt.plot(x1,y1)
sns.scatterplot(x=or_data['input_1'],y=or_data['input_2'],hue=or_data['output'],s=200)
```

```python
clf3.coef_
```

```python
clf3.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(xor_data.iloc[:,0:2].values,xor_data.iloc[:,-1].values, clf=clf3, legend=2)
```

So we can easily that they can't be divided or separable.

Hence, we say that Perceptron trick will fail on Non-linear data.
